```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import pandas as pd

# Make sure productDisplayName is clean
merged_df["productDisplayName"] = merged_df["productDisplayName"].astype(str).str.strip().str.lower()

# Create vector from productDisplayName
vectorizer = TfidfVectorizer(stop_words='english')
vector = vectorizer.fit_transform(merged_df["productDisplayName"])

# Build series: product name → index
series = pd.Series(data=merged_df.index, index=merged_df["articleType"])
```

```python
def movie_dec(name, n):
    name = name.strip().lower()

    # Find best match containing the name
    matches = series[series.index.str.contains(name, case=False)]
    
    if matches.empty:
        return [{"error": f"❌ '{name}' not found in productDisplayName list."}]

    idx = matches.iloc[0]  # take the first match
    dis = linear_kernel(vector[idx], vector)

    scores = pd.DataFrame({"score": dis.flatten()})
    scores = scores.sort_values("score", ascending=False)

    lst = []
    for i in range(1, n + 1):
        product_idx = scores.index[i]
        s = {
            "product name": merged_df["articleType"].iloc[product_idx],
            "subCategory": merged_df["subCategory"].iloc[product_idx],
            "baseColour": merged_df["baseColour"].iloc[product_idx],
            "productDisplayName": merged_df["productDisplayName"].iloc[product_idx],
            "img": merged_df["link"].iloc[product_idx],
            "gender": merged_df["gender"].iloc[product_idx]
        }
        lst.append(s)
    return lst
```

```python
name = input("Enter full product name: ")  # e.g., puma men grey t-shirt
n = int(input("How many similar items to show?: "))  # e.g., 5

output = movie_dec(name, n)
pd.DataFrame(output)
```
